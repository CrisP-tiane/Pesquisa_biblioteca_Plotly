# Biblioteca Plotly

_Construído sobre a biblioteca Plotly JavaScript ( plotly.js ), plotlypermite que os usuários do Python criem belas visualizações interativas baseadas na web que podem ser exibidas em notebooks Jupyter, salvas em arquivos HTML autônomos ou servidas como parte de aplicativos da web puros construídos em Python usando Dash. A plotlybiblioteca Python às vezes é chamada de "plotly.py" para diferenciá-la da biblioteca JavaScript._

## Importando outras bibliotecas

_Para uma melhor compreensão dos recurso da Plotly vamos comparar com a biblioteca Matplotlib._

# Gráficos


## Linha

```python
# Definação de arrays

x = np.arange(0.25, 20, 0.01) # Eixo horizontal

y1 = 12 * np.sin(x) / x       # Primeira curva >> 12 vezes o seno de x sobre x

y2 = 4 * np.log(x)            # Segunda curva >> função logaritma
```

```python
# Como ficaria com o Matplotlib?

plt.plot(x, y1, x, y2) # Plotando os dois gráficos
```

```python
# Importar um submódulo do Plotly chamado express como px

import plotly.express as px
```

```python
# E com o Plotly?

px.line(x=x, y=y1)
```

**Notaram a diferença?** 😉

```python
# Dois gráficos

px.line(x=x, y=[y1, y2]) # O y é colocado no array como uma lista
```

## Utilizando um DataFrame

```python
# Carregando um dataset
# Fonte: https://www.kaggle.com/datasets/kumarajarshi/life-expectancy-who

df = pd.read_csv('/content/Life Expectancy Data.csv')
```

```python
# Visualizando os dados

df
```

```python
df_brazil = df[df['Country'] == 'Brazil'] # Extraindo os dados somente de Brazil
```

```python
df_brazil
```

```python
# Com o Matplotlib

plt.plot(df_brazil['Year'], df_brazil['Life expectancy '])
```

```python
# Com o Plotly

px.line(df_brazil, x='Year', y='Life expectancy ', title="Gráfico de Linha", labels={
    'Year': 'Ano',
    'Life expectancy ': 'Expectativa de Vida'},
    )

#  title inclui um título; labels muda o nome das colunas usando a lógica de chave:valor; template altera o background do gráfico (por exemplo, ggplot2 ou gridon)
```

```python
# Vários gráficos

px.line(df, x='Year', y='Life expectancy ', color='Country')
```

## Dispersão

_O gráfico de dispersão representa valores para duas variáveis ​​numéricas diferentes . Eles são usados ​​principalmente para representação de relacionamento entre duas variáveis._

```python
# Definação de arrays

x = np.arange(0.25, 20, 1) # Eixo horizontal

y1 = 12 * np.sin(x) / x       # Primeira curva >> 12 vezes o seno de x sobre x

y2 = 4 * np.log(x)            # Segunda curva >> função logaritma
```

```python
# Plotly

px.scatter(x=x, y=[y1, y2]) # Scatter é a função para gráfico de dipsersão
```

```python
# Dispersão com o DataFrame
# Fonte: https://www.kaggle.com/datasets/kumarajarshi/life-expectancy-who

px.scatter(df, x='Year', y='Adult Mortality', color='Country', title = 'Título do Gráfico')
```

## Barras

_Os gráficos de barras são usados ​​quando queremos comparar diferentes grupos de dados e fazer inferências sobre quais grupos são mais altos e quais são comuns e comparar o desempenho de um grupo em comparação com outros._

```python
df_consumidor = pd.read_csv('/content/Customers.csv')
df_consumidor.head()
```

```python
df_profissao = df_consumidor.groupby('Profession')['Annual Income ($)'].mean().reset_index()
df_profissao
```

```python
# Gráfico de Barras

px.bar(df_profissao, x='Profession', y='Annual Income ($)')
```

### Outro exemplo de Gráfico de Barras

```python
# Gráfico de Barras com o DataFrame
# Fonte: https://www.kaggle.com/datasets/kumarajarshi/life-expectancy-who

px.bar(df, x='Year', y='Adult Mortality', color='Country', barmode='group') # Mudamos o gráfico mechendo nos parâmetros. Por exemplo barmode 
```

## Histograma

_Um histograma plota a distribuição contínua da variável como uma série de barras e cada barra indica a frequência do valor que ocorre em uma variável. Para usar um histograma, simplesmente exigimos uma variável que assume valores numéricos contínuos._

```python
px.histogram(df_consumidor, x='Age')
```

## Caixa (Boxplot)

_Um gráfico de caixa é a representação do resumo estatístico. Mínimo, primeiro quartil, mediana, terceiro quartil, máximo._

```python
px.box(df_consumidor, x='Age')
```

## Treemap

_Um mapa de árvore mostra os dados organizados em hierarquias de dimensão.Cada linha do DataFrame é representada como setor de um Mapa da Árvore_


```python
fig = px.treemap(
    names = ["Eve","Cain", "Seth", "Enos", "Noam", "Abel", "Awan", "Enoch", "Azura"],
    parents = ["", "Eve", "Eve", "Seth", "Seth", "Eve", "Eve", "Awan", "Eve"]
)
fig.update_traces(root_color="lightgrey")
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

### Treemap 2

```python
fig = px.treemap(df, path=['Country'], values='Life expectancy ',
                  color_continuous_scale='RdBu')
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

## Funil

```python
# Funil
import plotly.express as px

data = dict(
    metrica =[1000, 500, 250, 100, 45],
    etapa   =["Tráfego", "Vistantes", "Leads", "Oportunidades", "Clientes"])
fig = px.funnel(data, x='metrica', y='etapa')
fig.show()
```

## Plotagem de Superfície Topográfica 3D

```python
# Plotagem de Superfície Topográfica 3D

import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
